```python
import pandas as pd
import seaborn as sns
```

## Train CSV File

```python
train = pd.read_csv('train.csv', parse_dates=True, dayfirst=True, index_col="Date")
train.head()
```

```python
# Create a Month column
train["Month"] = train.index.month
```

```python
# Encode the StateHoliday column
train['StateHoliday'] = train['StateHoliday'].replace(["a", "b", "c"], [1, 2, 3])
train['StateHoliday'] = train['StateHoliday'].values.astype(int)
```

```python
# Customers does not need to be used, as it is not in test file
train = train.drop(['Customers'], axis=1)
```

```python
train.head()
```

## Test CSV File

```python
test = pd.read_csv('test.csv', parse_dates=True, dayfirst=True, index_col="Date")
test.head()
```

```python
# Create a Month column
test["Month"] = test.index.month
```

```python
# Encode the StateHoliday column
test['StateHoliday'] = test['StateHoliday'].replace(["a", "b", "c"], [1, 2, 3])
test['StateHoliday'] = test['StateHoliday'].values.astype(int)
```

```python
test['Open'] = test['Open'].fillna(0.0)
```

```python
test.head()
```

## Stores CSV File

```python
stores = pd.read_csv('store.csv')
stores.head()
```

```python
# Encode the StoreType column
stores['StoreType'] = stores['StoreType'].replace(["a", "b", "c", "d"], [1, 2, 3, 4])
```

```python
# Encode the Assortment column
stores['Assortment'] = stores['Assortment'].replace(["a", "b", "c"], [1, 2, 3])
```

```python
# Normalise the CompetitionDistance column
max = stores["CompetitionDistance"].max()
stores['CompetitionDistance'] = (stores['CompetitionDistance'] / max).values.astype(float)
```

```python
stores = stores.drop(['CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'Promo2SinceWeek', 'Promo2SinceYear', 'Promo2'], axis=1)
```

```python
stores["PromoInterval"] = stores["PromoInterval"].fillna("None")
stores = pd.get_dummies(stores, columns=['PromoInterval'])
stores.head()
```

```python
stores['PromoInterval_Feb,May,Aug,Nov'] = stores['PromoInterval_Feb,May,Aug,Nov'].replace([True, False], [1, 0])
stores['PromoInterval_Jan,Apr,Jul,Oct'] = stores['PromoInterval_Jan,Apr,Jul,Oct'].replace([True, False], [1, 0])
stores['PromoInterval_Mar,Jun,Sept,Dec'] = stores['PromoInterval_Mar,Jun,Sept,Dec'].replace([True, False], [1, 0])
stores['PromoInterval_None'] = stores['PromoInterval_None'].replace([True, False], [1, 0])
stores.head()
```

## Combining Dataframes

```python
training = pd.merge(left=train, right=stores, how="inner")
training.head()
```

```python
testing = pd.merge(left=test, right=stores, how="inner")
testing = testing.sort_values(by=['Id'])
testing.head()
```

```python
testing = testing.drop(['Id'], axis=1)
testing.head()
```

```python
training.to_csv("../rossmann-store-sales/training_data.csv")
testing.to_csv("../rossmann-store-sales/testing_data.csv")
```
